## Modular Arithmetic
$a$ is congruent to $b$ modulo $n$
$$ a \equiv b \pmod{n} $$
if $$a=b+kn$$ for some $k$

```python
# python 3 division
# float
print(4/7)
# integral (floor)
print(4//7)
# remainder
print(4%7)

# modulo equivalences
c = []
a = 3
n = 26
for k in range(-1,4):
    c.append(a+k*n)
    
print(f'{a} is equivalent to ...{c}... modulo {n}')
# double-check:
print( [j %26 for j in c] )
```

## Caesar Cipher
Circularly shift letters by a fixed amount.

```python
# alphabet: lowercase letters
from string import ascii_lowercase as alphabet

def caesar_encrypt(key: int, message: str) -> str:
    # initialise empty ciphertext
    ct = ''
    # iterate on every character
    for c in message:
        # substitute only alphabetical characters
        if c in alphabet:
            # original index of character
            i = alphabet.find(c)
            # shift according to key
            ct += alphabet[(i + key) % len(alphabet)]
        else:
            ct += c
    return ct

def caesar_decrypt(key: int, ciphertext: str) -> str:
    # to decrypt it suffices to subtract the key instead of adding it
    return caesar_encrypt(-key, ciphertext)

m = 'hello world'
k = 3
e = caesar_encrypt(k, m)
d = caesar_decrypt(k, e)
print(e)
print(d)
```

## Substitution Cipher
The key is a random permutation of the alphabet

```python
# the key: a permutation of the alphabet
key = 'qwertyuiopasdfghjklzxcvbnm'
# check that it is a permutation:
assert len(key) == len(alphabet)
for c in key:
    assert c in alphabet

# base function: substitute characters according to a permutation
def substitute(string: str, original: str, permuted: str) -> str:
    # initialise empty string
    result = ''
    # iterate on every character
    for c in string:
        # substitute only alphabetical characters
        if c in original:
            # find char in original order and return same position in the permuted
            result += permuted[original.find(c)]
        else:
            result += c
    return result

def sub_encrypt(key: str, message: str) -> str:
    # use alphabet as original and key as permutation
    return substitute(message, alphabet, key)

def sub_decrypt(key: int, ciphertext: str) -> str:
    # to decrypt it suffices to swap key and alphabet
    return substitute(ciphertext, key, alphabet)

m = 'hello world'
e = sub_encrypt(key, m)
d = sub_decrypt(key, e)
print(e)
print(d)
```

```python
from collections import Counter
#frequency analysis of a text
def frequencies(text: str):
    # initialise counters
    letter_freq = Counter()
    bigram_freq = Counter()
    trigram_freq = Counter()
    # analyse text up to third to last character
    n = len(text)
    for i in range(n - 2):
        letter_freq[text[i]] += 1
        bigram_freq[text[i: i + 2]] += 1
        trigram_freq[text[i: i + 3]] += 1
    # count last letters
    letter_freq[text[n - 2]] += 1
    bigram_freq[text[n - 2: n - 1]] += 1
    letter_freq[text[n - 1]] += 1
    print('Most frequent letters:', letter_freq.most_common(11))
    print('Most frequent bigrams:', bigram_freq.most_common(9))
    print('Most frequent trigrams:', trigram_freq.most_common(8))

txt = 'ahidaygboajxibzdxybaiayahgayjliacoqjmhvbhabhiimcbzebjaybxhgahbvaixcoqgavagbidbaewoxhuyidbebwmebidbmegbeyxebimxiixjuxigxfhahidbpaggobmwidbaewmepxiamhidbjmhibhimwidbpbyyxrbdxyimebpxahyxwbxhgzemibjibgwemplhxlidmeaybgxjjbyyxhgymzobxybcbjxebwloxhggmhmigavlorxibidbzebjamlyahwmepxiamhyjmhixahbgdbebahidxijmlogcbdbozwlowmeidbbhbpq'

frequencies(txt)
```
